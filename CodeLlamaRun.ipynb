```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch
```

```python
def text_generation(tokenizer, generated_ids, prompt):
  return(tokenizer.batch_decode(generated_ids)[0].replace("<s>","").strip())
```

```python
def generate_code(prompt, model, tokenizer):
  model_inputs = tokenizer(prompt, return_tensors ="pt", add_special_tokens = True).to("cuda")
  generated_ids = model.generate(**model_inputs, do_sample = True,
                                 pad_token_id = tokenizer.eos_token_id,
                                 temperature = 0.1,
                                 top_p = 0.9,
                                 num_return_sequences = 1,
                                 eos_token_id = tokenizer.eos_token_id,
                                 max_length = 128)
  return(text_generation(tokenizer, generated_ids, prompt))
```

```python
model_id = "codellama/CodeLlama-7b-hf"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype = torch.float32, load_in_4bit = True)
```

```python
print(generate_code("def array insertion(", model, tokenizer))
```

```python
print(generate_code("def fibonacci(", model, tokenizer))
```

```python
print(generate_code("def is_perfect_number(", model, tokenizer))
```

```python
#fibonacci
def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))
```
